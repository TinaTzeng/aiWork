```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd 
import jieba
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
url = "https://www.ptt.cc/bbs/nCoV2019/index.html"
title = []
# 爬ptt武漢肺炎討論版三頁
for i in range(3):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    sel = soup.select("div.title a")
    u = soup.select("div.btn-group.btn-group-paging a")
    print(url)
    url = "https://www.ptt.cc" + u[1]["href"]
    
    for s in sel:
        print(s.text)
        title.append(s.text)
```

```python
jieba.add_word('安倍晉三', freq=None, tag=None)
jieba.add_word('義大利', freq=None, tag=None)
cutL = []
for t in title:
    qseg_list = jieba.cut(t, cut_all = False)
    cutL.append(list(qseg_list))
```

```python
text = ''
for se in cutL:
    for wd in se:
        text = text + wd + ','
```

```python

stopwords = {}.fromkeys(["問題","情報","討論","建議","新聞","Re"])

wc = WordCloud(font_path="NotoSansCJKtc-Black.otf",
               background_color="white", #背景顏色
               max_words = 2000 ,        #文字雲顯示最大詞數
               stopwords=stopwords)      #停用字詞

# 直接使用結巴斷詞後的文字來產生文字雲
wc.generate_from_text(text)
# wc = WordCloud().generate(text)
 
# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(20,10), dpi =200)
plt.show()
```
