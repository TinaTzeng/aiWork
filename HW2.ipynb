```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
```

```python
Options.binary_location = "C:\\Program Files (x86)\\Google\\Chrome\\Application\\chrome.exe"
webdriver_path = 'D:\\chromedriver.exe'
options = Options()
driver = webdriver.Chrome(executable_path=webdriver_path, options=options)
driver.get("https://stats.nba.com/players/traditional/?sort=PTS&dir=-1")
```

```python
a = driver.find_elements_by_class_name('nba-stat-table__overflow')
# a = [PLAYER........., ] (只有一個值)
for data in a:
    print(data.text)
    d = data.text.split("\n")
```

```python
columns = d[0]
# print(type(columns)) #columns is string type
columns = columns.split(" ") 
print(columns) #type(columns) = list
```

```python
#抓出player與其他資料
print(d[2:4]) #第一列
#找出最後一列的index
print(2+3*49)
print(d[149:151]) #最後一列
```

```python
#抓出所有列的資料
rows = []
i = 2
while i <= 149:
    print(d[i:i+2])
    rows.append(d[i:i+2])
    i += 3
```

```python
#把每筆資料切成跟欄位一樣的數量(29)
for i in range(len(rows)):
    rows[i][1] = rows[i][1].split(' ')
    for j in range(1,29):
        rows[i].insert(j, rows[i][j][j-1])
    del rows[i][-1]
```

```python
#做成表格
import pandas as pd
df = pd.DataFrame(data = rows, columns = columns)
df.index.name = 'RANK'
df.index = df.index + 1
df
```
