```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data = pd.read_csv('happiness/2019.csv')
data
```

```python
# 找出前15名最快樂的國家/地區
happiest = data.head(15)
happiest.plot.bar(x='Country or region', y='Score', color = 'orange')
```

```python
happiest
# 人均國內生產毛額（Log GDP per capita）
# 社會支持資源（Social support）-- 評估遇到困難時支援系統的可得性
# 預期健康壽命（Healthy life expectancy）
# 社會自由度（Freedom to make life choices）
# 對他人慷慨程度（Generosity）-- 透過捐贈來評估程度
# 對政府的信任度，以貪腐程度衡量（Perceptions of corruption）
```

```python
# 看線性相關性
corrMatrix = happiest.corr()
sn.heatmap(corrMatrix, annot=True)
#Score跟Social Support相關性最高(0.68)
#第二高為Freedom to make life choices社會自由度(0.66)
#可得知2019世界最幸福的前15個國家人民感受到幸福的主要因素跟國家的社會系統有關
```
